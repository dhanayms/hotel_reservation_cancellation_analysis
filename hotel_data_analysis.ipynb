```python
#romoving columns with large number of null values
df.drop(['company','agent'], axis = 1, inplace = True)
df.dropna(inplace = True)
```

```python
df['adr'].plot(kind = 'box')
```

```python
df=df[df['adr']<5000]
df['adr'].plot(kind = 'box')
```

```python
cancel_percentage = df['is_canceled'].value_counts(normalize=True)
cancel_percentage
```

```python
plt.figure(figsize = (5,5))
plt.title('Reservation status count')
plt.bar(['notcancelled','cancelled'],df['is_canceled'].value_counts() ,edgecolor = 'k',width = 0.5)
```

```python
plt.figure(figsize=(7,7)) 
ax1=sns.countplot(x = 'hotel',hue = 'is_canceled' , data = df )
plt.legend(title='Staus', loc='upper left', labels=['not cancelled', 'cancelled'])
plt.title('Reservation status in TWo hotels', size=20)
plt.xlabel('Hotel')
plt.ylabel('No of reservations')
plt.show()
```

```python
resort_hotel=df[df['hotel']== 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']== 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,7))
plt.title('average daily rate',fontsize=20)
plt.plot(resort_hotel.index,resort_hotel['adr'],label = 'resort_hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label = 'city_hotel')
plt.legend()
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(20,7))
ax1=sns.countplot(x='month',hue='is_canceled' ,data=df)
plt.legend(['not cancelled', 'cancelled'])
plt.title('Reservation status in TWo hotels', size=20)
plt.xlabel('month')
plt.ylabel('No of reservations')
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title('Rate per month')
sns.barplot('month','adr',data=df[df['is_canceled'] == 1].groupby('month')['adr'].sum().reset_index())
#plt.legend(fontsize=10)
plt.show()
```

```python
cancelled = df[df['is_canceled'] == 1]
top_country=cancelled['country'].value_counts()[:10]
plt.figure(figsize=(7,7))
plt.title('top 10 countries')
plt.pie(top_country , autopct = '%.2f' , labels = top_country.index)
plt.show()
```

```python
df['market_segment'].value_counts(normalize=True)
```

```python
cancelled['market_segment'].value_counts(normalize=True)
```

```python
not_cancelled = df[df['is_canceled'] == 0]
```

```python
rate_of_cancelled=cancelled.groupby('reservation_status_date')[['adr']].mean()
rate_of_cancelled.reset_index(inplace=True)
rate_of_cancelled.sort_values('reservation_status_date',inplace=True)


rate_of_notcancelled=not_cancelled.groupby('reservation_status_date')[['adr']].mean()
rate_of_notcancelled.reset_index(inplace=True)
rate_of_notcancelled.sort_values('reservation_status_date',inplace=True)


plt.figure(figsize=(15,5))
plt.title('Average daily rate')
plt.plot(rate_of_cancelled['reservation_status_date'],rate_of_cancelled['adr'],label='cancelled')
plt.plot(rate_of_notcancelled['reservation_status_date'],rate_of_notcancelled['adr'],label='cancelled')
plt.legend()
```

```python
rate_of_cancelled=rate_of_cancelled[(rate_of_cancelled['reservation_status_date'] > '2016') &(rate_of_cancelled['reservation_status_date'] <'2017-09')]


rate_of_notcancelled=rate_of_notcancelled[(rate_of_notcancelled['reservation_status_date'] > '2016') &(rate_of_notcancelled['reservation_status_date'] <'2017-09')]
```

```python
plt.figure(figsize=(15,6))
plt.title('Average daily rate',fontsize=22)
plt.plot(rate_of_cancelled['reservation_status_date'],rate_of_cancelled['adr'],label='cancelled')
plt.plot(rate_of_notcancelled['reservation_status_date'],rate_of_notcancelled['adr'],label='not cancelled')
plt.legend()
```
